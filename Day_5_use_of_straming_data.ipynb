```python
import numpy as np
import time
import pickle
import pandas as pd

# 🔹 Load tuned model (only once)
with open("/content/rf_model.pkl", "rb") as f:
    model = pickle.load(f)
```

```python

# 🔹 Load dataset
df = pd.read_csv("ml_ready_iot_data.csv")
df.columns = df.columns.str.strip()

# Feature order
FEATURES = [
    "temperature",
    "humidity",
    "usage_hours",
    "error_count",
    "vibration",
    "avg_temp",
    "temp_change",
    "usage_frequency",
    "health_score"
]
```

```python
#Example stram data
stream_data = [
    {
        "device_id": "D101",
        "temperature": 60, "humidity": 75, "usage_hours": 12,
        "error_count": 2, "vibration": 0.6,
        "avg_temp": 60, "temp_change": 5,
        "usage_frequency": 0.5, "health_score": 55
    },
    {
        "device_id": "D102",
        "temperature": 30, "humidity": 50, "usage_hours": 5,
        "error_count": 0, "vibration": 0.2,
        "avg_temp": 30, "temp_change": 2,
        "usage_frequency": 0.2, "health_score": 80
    }
]
```

```python
print("Streaming Started...\n")

# Stream loop
for index, row in df.iterrows():

    # Prepare input
    input_data = np.array([[row[f] for f in FEATURES]])

    # Prediction with error handling
    try:
        status = model.predict(input_data)[0]
    except Exception as e:
        status = "error"

    # Output
    print({
        "device_id": row["device_id"],
        "status": status
    })

    time.sleep(1)
```
